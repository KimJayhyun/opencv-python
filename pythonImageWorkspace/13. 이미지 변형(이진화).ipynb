## Adaptive Threshold
이미지를 작은 영역으로 나누어서 임계치 적용

```python
import cv2
path = "C:\\Users\\User\\Desktop\\code\\gitFolder\\myStudy\\3_OpenCV_python\\pythonImageWorkspace\\Data"

def empty(pos):
    # print(pos)
    pass

img = cv2.imread(path + '/book.jpg', cv2.IMREAD_GRAYSCALE)

name = 'Trackbar'
cv2.namedWindow(name)

# bar 이름, 창의 이름, 초기값, 최대값, 이벤트 처리
cv2.createTrackbar('block_size', name, 25, 100, empty) # 홀수만 가능, 1보다 큰 값
cv2.createTrackbar('c', name, 3, 10, empty) # 일반적으로 양수의 값을 사용

while True:
    # bar 이름, 창의 이름
    block_size = cv2.getTrackbarPos('block_size', name)
    c = cv2.getTrackbarPos('c', name)

    if block_size <= 1:
        block_size = 3
    
    if block_size % 2 == 0:
        block_size += 1


    binary = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, c)
    
    cv2.imshow(name, binary)
    
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

## 오츠 알고리즘
Bimodal Image에 사용하기 적합
최적의 임계치를 자동으로 계산

```python
import cv2
path = "C:\\Users\\User\\Desktop\\code\\gitFolder\\myStudy\\3_OpenCV_python\\pythonImageWorkspace\\Data"

img = cv2.imread(path + '/book.jpg', cv2.IMREAD_GRAYSCALE)

ret, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
ret, otsu = cv2.threshold(img, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

print('otsu threshold', ret)

cv2.imshow('title', img)
cv2.imshow('binary', binary)
cv2.imshow('otsu', otsu)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
